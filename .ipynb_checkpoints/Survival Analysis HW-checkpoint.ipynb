# Homework 3 | Survival Analysis & CLV

Survival analysis on a telecom subscriber dataset using parametric AFT models, model selection, CLV calculation, and segment exploration.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from lifelines import (
    WeibullAFTFitter,
    LogNormalAFTFitter,
    LogLogisticAFTFitter,
    KaplanMeierFitter
)

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (20, 8)
```

## 1. Data Loading & Preparation

```python
df = pd.read_csv('telco.csv')
print(f"Shape: {df.shape}")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df_encoded = df.copy()

binary_maps = {
    'marital':  {'Unmarried': 0, 'Married': 1},
    'retire':   {'No': 0, 'Yes': 1},
    'gender':   {'Male': 0, 'Female': 1},
    'voice':    {'No': 0, 'Yes': 1},
    'internet': {'No': 0, 'Yes': 1},
    'forward':  {'No': 0, 'Yes': 1}
}

for col, mapping in binary_maps.items():
    if col in df_encoded.columns and df_encoded[col].dtype == 'object':
        df_encoded[col] = df_encoded[col].map(mapping)

for col in ['ed', 'custcat', 'region']:
    if col in df_encoded.columns and df_encoded[col].dtype == 'object':
        df_encoded = pd.get_dummies(df_encoded, columns=[col], drop_first=True, dtype=int)

if 'ID' in df_encoded.columns:
    df_encoded = df_encoded.drop(columns=['ID'])

df_encoded.head()
```

## 2. Kaplan-Meier Baseline

```python
df_encoded['churn'] = df_encoded['churn'].map({'Yes': 1, 'No': 0})

kmf = KaplanMeierFitter()
kmf.fit(df_encoded['tenure'], event_observed=df_encoded['churn'], label='KM Estimate')

fig, ax = plt.subplots()
kmf.plot_survival_function(ax=ax)
ax.set_title('Kaplan-Meier Survival Curve')
ax.set_xlabel('Tenure (months)')
ax.set_ylabel('Survival Probability')
plt.tight_layout()
plt.show()

print(f"Median survival time: {kmf.median_survival_time_:.2f} months")
```

## 3. Parametric AFT Models

Fitting Weibull, Log-Normal, and Log-Logistic AFT models with all covariates.

```python
aft_models = {'Weibull':      WeibullAFTFitter(penalizer=0.01),
              'Log-Normal':   LogNormalAFTFitter(penalizer=0.01),
              'Log-Logistic': LogLogisticAFTFitter(penalizer=0.01)}

fitted_models = {}
results = []

for name, model in aft_models.items():
    model.fit(df_encoded, duration_col='tenure', event_col='churn')
    fitted_models[name] = model
    results.append({'Model': name,
                    'AIC': model.AIC_,
                    'BIC': model.BIC_,
                    'Log-Likelihood': model.log_likelihood_,
                    'Concordance': model.concordance_index_})
    print(f"\n{'='*60}\n{name} AFT\n{'='*60}")
    model.print_summary(columns=['coef', 'exp(coef)', 'p'])
```

### Model Comparison

```python
comparison = pd.DataFrame(results).set_index('Model')
comparison
```

### Survival Curves — All Models on One Plot

```python
fig, ax = plt.subplots(figsize=(20, 8))
kmf.plot_survival_function(ax=ax, ci_show=False, color='black', linewidth=2)

colors = {'Weibull': '#e74c3c', 'Log-Normal': '#3498db', 'Log-Logistic': '#2ecc71'}
times = np.linspace(0.5, df_encoded['tenure'].max(), 200)

for name, model in fitted_models.items():
    surv = model.predict_survival_function(df_encoded, times=times).mean(axis=1)
    ax.plot(times, surv, label=f'{name} AFT', color=colors[name], linewidth=1.5)

ax.set_title('Survival Curves: KM vs Parametric AFT Models')
ax.set_xlabel('Tenure (months)')
ax.set_ylabel('Survival Probability')
ax.legend()
plt.tight_layout()
plt.show()
```

### Model Selection

We select the model with the lowest AIC. Beyond AIC, the Log-Logistic distribution allows non-monotone (hump-shaped) hazards, which is realistic in churn: risk may first rise then fall as surviving customers become increasingly loyal.

```python
best_name = comparison['AIC'].idxmin()
print(f"Best model by AIC: {best_name}")
best_model = fitted_models[best_name]
```

## 4. Feature Selection — Significant Features Only

Refitting with only p < 0.05 features.

```python
summary = best_model.summary
sig_mask = summary['p'] < 0.05
sig_features = [idx[1] for idx in summary[sig_mask].index if idx[1] not in ['Intercept', '_intercept']]
sig_features = list(set(sig_features))
print(f"Significant features: {sig_features}")

keep_cols = sig_features + ['tenure', 'churn']
keep_cols = [c for c in keep_cols if c in df_encoded.columns]
df_final = df_encoded[keep_cols].copy()
```

```python
model_classes = {
    'Weibull':      WeibullAFTFitter,
    'Log-Normal':   LogNormalAFTFitter,
    'Log-Logistic': LogLogisticAFTFitter
}

final_model = model_classes[best_name](penalizer=0.01)
final_model.fit(df_final, duration_col='tenure', event_col='churn')
final_model.print_summary(columns=['coef', 'exp(coef)', 'p'])
print(f"\nFinal Model AIC: {final_model.AIC_:.2f}")
print(f"Final Model Concordance: {final_model.concordance_index_:.4f}")
```

## 5. CLV Calculation

Using the CLV formula: $CLV = \sum_{t=1}^{T} \frac{m \cdot S(t)}{(1+d)^t}$, where $S(t)$ is the predicted survival probability, $m$ is monthly margin, and $d$ is the monthly discount rate.

```python
MONTHLY_REVENUE = 50
MONTHLY_COST = 30
MONTHLY_MARGIN = MONTHLY_REVENUE - MONTHLY_COST
ANNUAL_DISCOUNT = 0.10
MONTHLY_DISCOUNT = ANNUAL_DISCOUNT / 12
T_HORIZON = 72

times = np.arange(1, T_HORIZON + 1)
surv_probs = final_model.predict_survival_function(df_final, times=times)
discount_factors = np.array([1 / (1 + MONTHLY_DISCOUNT) ** t for t in times])

clv_per_customer = (surv_probs.values * MONTHLY_MARGIN * discount_factors.reshape(-1, 1)).sum(axis=0)
df['CLV'] = clv_per_customer

print("CLV Statistics:")
print(df['CLV'].describe())
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
ax.hist(df['CLV'], bins=40, edgecolor='white', color='#3498db', alpha=0.8)
ax.axvline(df['CLV'].mean(), color='red', linestyle='--', label=f"Mean: ${df['CLV'].mean():.0f}")
ax.axvline(df['CLV'].median(), color='orange', linestyle='--', label=f"Median: ${df['CLV'].median():.0f}")
ax.set_title('Distribution of Customer Lifetime Value')
ax.set_xlabel('CLV ($)')
ax.set_ylabel('Frequency')
ax.legend()
plt.tight_layout()
plt.show()
```

## 6. CLV by Segments

```python
segment_cols = [c for c in ['gender', 'marital', 'custcat', 'internet', 'voice', 'retire'] if c in df.columns]

for col in segment_cols:
    print(f"\n--- CLV by {col} ---")
    print(df.groupby(col)['CLV'].agg(['mean', 'median', 'count']).round(2))
```

```python
if 'custcat' in df.columns:
    fig, ax = plt.subplots(figsize=(10, 5))
    df.boxplot(column='CLV', by='custcat', ax=ax)
    ax.set_title('CLV by Customer Category')
    ax.set_xlabel('Customer Category')
    ax.set_ylabel('CLV ($)')
    plt.suptitle('')
    plt.tight_layout()
    plt.show()
```

```python
df['clv_segment'] = pd.qcut(df['CLV'], q=4, labels=['Low', 'Medium', 'High', 'VIP'])
print(df.groupby('clv_segment')['CLV'].agg(['mean', 'median', 'count', 'sum']).round(2))
```

## 7. Retention Budget Estimation

Identifying at-risk subscribers (>50% conditional churn probability within 12 months) and sizing the annual retention budget.

```python
current_tenure = df_final['tenure'].values

surv_now_vals = []
surv_fut_vals = []

for i in range(len(df_final)):
    row = df_final.iloc[[i]]
    t_now = max(current_tenure[i], 1)
    t_fut = current_tenure[i] + 12
    s_now = final_model.predict_survival_function(row, times=[t_now]).values[0, 0]
    s_fut = final_model.predict_survival_function(row, times=[t_fut]).values[0, 0]
    surv_now_vals.append(s_now)
    surv_fut_vals.append(s_fut)

df['surv_now'] = surv_now_vals
df['surv_12m'] = surv_fut_vals
df['cond_surv_12m'] = df['surv_12m'] / df['surv_now']
df['churn_prob_12m'] = 1 - df['cond_surv_12m']

at_risk = df[df['churn_prob_12m'] > 0.5]
at_risk_clv = at_risk['CLV'].sum()
retention_budget = at_risk_clv * 0.10

print(f"At-risk subscribers (>50% churn prob in 12m): {len(at_risk)}")
print(f"Total CLV of at-risk subscribers: ${at_risk_clv:,.0f}")
print(f"Annual retention budget (10% of at-risk CLV): ${retention_budget:,.0f}")
```
